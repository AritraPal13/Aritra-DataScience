**Q6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
def get_median(sample):                                #created a user defined(get_median) funtion to get median of any data.
     n = len(sample)                                   #len-length of the data set  //-Floor Division  %-Modulus operator
     index = n // 2
     # Sample with an odd number of observations
     if n % 2:
         return sorted(sample)[index]
     # Sample with an even number of observations
     return sum(sorted(sample)[index - 1:index + 1]) / 2
```

```python
get_median(housing['longitude'])     #getting the median by user defined function
```

```python
housing['longitude'].median()     #getting median by using the median() function
```

**Q10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
#conditions
case = [(housing['total_bedrooms'] <=10),                                        
        (housing['total_bedrooms'] >=11)&(housing['total_bedrooms'] <=1000),
        (housing['total_bedrooms'] > 1000)]
#choicelist
size = ['small','medium','large']                                                #Syntax : numpy.select(condlist, choicelist, default = 0)
#numpy.select() function
housing['total_bedroom_size']=np.select(case,size)                              
#numpy.select() function return an array drawn from elements in choicelist, depending on conditions.
#output
housing                              
```

We have created a new column called 'total_bedroom_size' which is based on the number of total_bedrooms.

If 

total_bedrooms<=10 then total_bedroom_size=small

total_bedrooms>=11 & total_bedrooms<=1000 then total_bedroom_size=medium

total_bedrooms>1000 then total_bedroom_size=large
